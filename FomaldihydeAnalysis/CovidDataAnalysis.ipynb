```python
import pandas as pd
import requests

country = 'sri lanka'
api_url = 'https://api.api-ninjas.com/v1/covid19?country={}'.format(country)

headers = {'X-Api-Key': '3EUMbhcelIreko7nxnShRw==6LOVjUhGnyFbocX7'}

try:
    response = requests.get(api_url, headers=headers)
    
    if response.status_code == requests.codes.ok:
        data = response.json()
        
        # Extracting relevant data
        country_data = data
        
        # Creating DataFrame
        df = pd.DataFrame(country_data)
        
    else:
        print("Error:", response.status_code, response.text)
        
except Exception as e:
    print("Error:", e)
```

```python
df.head()
```

```python
sl_data = pd.DataFrame.from_dict(df["cases"][0], orient='index')
sl_data.index.name = 'date'


sl_data = sl_data[['total', 'new']]
sl_data.reset_index(inplace=True)
```

```python
sl_data["country"] = ["Sri Lanka" for x in range(0, len(sl_data))]
```

```python
sl_data.head()
```

```python
sl_data = sl_data[["country", "date", "new", "total"]]
sl_data.head()
```

```python
sl_data.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert date to datetime
sl_data['date'] = pd.to_datetime(sl_data['date'])

# Set style
sns.set_style("whitegrid")

# Plot 1: Line plot of total cases over time
plt.figure(figsize=(8, 5))
sns.lineplot(data=sl_data, x='date', y='total', marker='o', color='blue', label='Total Cases')
plt.title('Total Cases Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Cases', fontsize=14)
plt.xticks(rotation=45)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

# Plot 2: Line plot of new cases over time
plt.figure(figsize=(8, 5))
sns.lineplot(data=sl_data, x='date', y='new', marker='o', color='green', label='New Cases')
plt.title('New Cases Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('New Cases', fontsize=14)
plt.xticks(rotation=45)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
