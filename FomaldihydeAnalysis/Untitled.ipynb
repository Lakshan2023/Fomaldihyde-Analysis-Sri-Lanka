## Combine All dataframes

```python
# Creating the "Region" column based on each DataFrame

colomboProperData['Region'] = ["Colombo Proper" for x in range(0, len(colomboProperData))]
mataraDeniyayaData['Region'] = ["Deniyaya, Matara" for x in range(0, len(mataraDeniyayaData))]
nuwaraEliyaData['Region'] = ["Nuwara Eliya Proper" for x in range(0, len(nuwaraEliyaData))]
bibileMonaragalaData['Region'] = ["Bibile, Monaragala" for x in range(0, len(bibileMonaragalaData))]
kurunegalaData['Region'] = ["Kurunegala Proper" for x in range(0, len(kurunegalaData))]
jaffnaData['Region'] = ["Jaffna Proper" for x in range(0, len(jaffnaData))]
kandyData['Region'] = ["Kandy Proper" for x in range(0, len(kandyData))]
```

```python
# Set the "Region" column as the index for each DataFrame
colomboProper.set_index('Region', inplace=True)
mataraDeniyayaData.set_index('Region', inplace=True)
nuwaraEliyaData.set_index('Region', inplace=True)
bibileMonaragalaData.set_index('Region', inplace=True)
kurunegalaData.set_index('Region', inplace=True)
jaffnaData.set_index('Region', inplace=True)
kandyData.set_index('Region', inplace=True)

# Concatenate DataFrames along the index axis
joined_data = pd.concat([colomboProper, mataraDeniyayaData, nuwaraEliyaData,
                         bibileMonaragalaData, kurunegalaData, jaffnaData, kandyData])

# Reset the index to make "Region" a column again
joined_data.reset_index(inplace=True)
```

```python
joined_data.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
grouped = joined_data.groupby("Region")
ax = None
for region, data in grouped:
    ax = data.plot(x='Next Date', y='HCHO reading', label=region, ax=ax, figsize=(10, 6), marker='o')

# Adding labels and title
plt.title('HCHO Readings of All Regions Over Years')
plt.xlabel('Year')
plt.ylabel('HCHO Reading')

# Adding legend
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objs as go
import pandas as pd

# Assuming joined_data is your DataFrame with regions as columns and years as rows
# You may need to pivot your data or prepare it differently based on your DataFrame structure

# Create traces for each region
traces = []
for region in joined_data['Region'].unique():
    region_data = joined_data[joined_data['Region'] == region]
    trace = go.Scatter(x=region_data['Next Date'], y=region_data['HCHO reading'], mode='lines+markers', name=region)
    traces.append(trace)

# Create buttons for each region
buttons = []
for i, region in enumerate(joined_data['Region'].unique()):
    visible = [False] * len(joined_data['Region'].unique())
    visible[i] = True
    button = dict(label=region, method='update', args=[{'visible': visible}, {'title': f'HCHO Readings of {region} Over Years'}])
    buttons.append(button)

# Layout settings
layout = dict(title='HCHO Readings of All Regions Over Years', xaxis=dict(title='Year'), yaxis=dict(title='HCHO Reading'),
              updatemenus=[dict(buttons=buttons, direction='down', x=0.01, xanchor='left', y=1.15, yanchor='top')])

# Create figure
fig = go.Figure(data=traces, layout=layout)

# Display the plot
fig.show()
```
